# Leitura e configuração do ambiente

```python
import pandas as pd
import geopandas
import math

csv_git_url = "https://raw.githubusercontent.com/PaixaoThales/GHI-VIS/main/VIS_Pr_03_hungry_in_the_world.csv"

data = pd.read_csv(csv_git_url, sep=",", encoding = "UTF-8")

def normalize_ghi(data: str):
  if data.__contains__("-"):
    return float(data.split("-")[1])
  elif data.__contains__("<"):
    return float(data.split("<")[1])
  elif data.__contains__("ﾖ"):
    return float(data.split("ﾖ")[1])
  return float(data)

years = ["2000", "2007", "2014", "2022"]
for year in years:
  data[year] = data[year].apply(lambda ghi: normalize_ghi(ghi))

def normalize(value: float):
  if(value == "Infinite" or not math.isfinite(value)):
    return 0
  return float(value)

data["ACC(%) 2000-2007"] = (((data["2007"] / data["2000"]) - 1) * 100).apply(lambda value: normalize(value))
data["ACC(%) 2007-2014"] = (((data["2014"] / data["2007"]) - 1) * 100).apply(lambda value: normalize(value))
data["ACC(%) 2014-2022"] = (((data["2022"] / data["2014"]) - 1) * 100).apply(lambda value: normalize(value))
data["ACC(%) 2000-2022"] = (((data["2022"] / data["2000"]) - 1) * 100).apply(lambda value: normalize(value))
del data["Absolute change since 2014"]
del data["Percent change sice 2014"]

data
```

# Fome Mundial Ranking

```python
import folium 

world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

data = world.merge(data, how="inner", left_on=["name"], right_on=["Country"])
data
```

# (%) GHI Acumulado 2000-2007 

```python

  map = folium.Map()

  folium.Choropleth(
      geo_data=data,
      name='choropleth',
      data=data,
      columns=['Country', "ACC(%) 2000-2007"],
      key_on='feature.properties.name',
      fill_color='OrRd',
      fill_opacity=0.7,
      line_opacity=0.2,
      legend_name='Percentual Acumulado GHI',
      zoom_start=2
  ).add_to(map)

map
```
